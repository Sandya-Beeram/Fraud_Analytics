```python
import networkx as nx

# Construct Graph
def construct_graph(data):
    G = nx.MultiDiGraph()
    for index, row in data.iterrows():
        G.add_weighted_edges_from([(row['Sender'], row['Receiver'], row['Amount'])])
    return G

myGraph = construct_graph(df)
```

```python
# GCN Model
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.utils import from_networkx
```

```python
# Convert the graph to PyTorch Geometric's data format
data = from_networkx(myGraph)
```

```python
edge_features = torch.tensor([myGraph[u][v][0]['weight'] for u, v in myGraph.edges()], dtype=torch.float)
```

```python
# Assuming each node's feature is a one-hot encoding of sender/receiver
num_nodes = len(myGraph.nodes)
node_features = torch.zeros(num_nodes, 2)  # Assuming there are only 2 categories: sender and receiver
for i, node in enumerate(myGraph.nodes):
    if myGraph.in_degree(node) > 0:  # Receiver
        node_features[i][1] = 1
    else:  # Sender
        node_features[i][0] = 1
```

```python
# Assign node features and edge features to the data object
data.x = node_features
data.edge_attr = edge_features.view(-1, 1)  # Reshape edge features to be compatible with PyTorch Geometric
```

```python
# Define GCN model
class GCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, data):
        x, edge_index, edge_attr = data.x, data.edge_index, data.edge_attr
        x = self.conv1(x, edge_index, edge_attr)
        x = F.relu(x)
        x = self.conv2(x, edge_index, edge_attr)
        return F.log_softmax(x, dim=1)
```

```python

from torch_geometric.nn import GCNConv

# Initialize GCN model
model = GCN(in_channels=2, hidden_channels=64, out_channels=32)
```

```python
output = model(data)
```

```python
from sklearn.cluster import KMeans


# Convert output tensor to numpy array
embeddings = output.detach().numpy()

# Apply K-means clustering
num_clusters = 3  # Number of clusters
kmeans = KMeans(n_clusters=num_clusters)
clusters = kmeans.fit_predict(embeddings)
```

```python
# Assign cluster labels to nodes in the graph
for i, node in enumerate(myGraph.nodes):
    myGraph.nodes[node]['cluster'] = clusters[i]
```

```python
import matplotlib.pyplot as plt

# Visualize the clusters in the graph
def visualize_clusters(graph):
    pos = nx.spring_layout(graph)  # Position nodes using Fruchterman-Reingold force-directed algorithm
    colors = ['r', 'g', 'b', 'y', 'c', 'm']  # Define colors for clusters (adjust as needed)
    
    plt.figure(figsize=(10, 6))
    for cluster_id in range(num_clusters):
        nodes_in_cluster = [node for node, data in graph.nodes(data=True) if data['cluster'] == cluster_id]
        nx.draw_networkx_nodes(graph, pos, nodelist=nodes_in_cluster, node_color=colors[cluster_id], label=f'Cluster {cluster_id}')
        
    nx.draw_networkx_edges(graph, pos, alpha=0.5)
    plt.title('Graph with Clusters')
    plt.legend()
    plt.axis('off')
    plt.show()

# Visualize clusters in the graph
visualize_clusters(myGraph)
```
